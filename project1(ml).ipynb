```python
y=data['status']
x=data.drop(['status'],axis=1)
```

```python
x.shape
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
	x, y, test_size=0.2, random_state=42)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)

# Create and train the logistic regression model
model = LogisticRegression()
model.fit(X_train_scaled, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test_scaled)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

# Display results
print(f'Accuracy: {accuracy:.2f}')
print(f'Confusion Matrix:\n{conf_matrix}')
print(f'Classification Report:\n{classification_rep}')
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
data_g=GaussianNB()
data_g.fit(x_train,y_train)
pre=data_g.predict(x_test)
pre=data_g.predict(x_test)
acc=accuracy_score(y_test,pre)
con=confusion_matrix(y_test,pre)
cla=classification_report(y_test,pre)

print(acc)
print(con)
print(cla)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
classifier.fit(x_train, y_train)
pre=classifier.predict(x_test)
cla=classification_report(y_test,pre)
acc=accuracy_score(y_test,pre)
con=confusion_matrix(y_test,pre)

print(acc)
print(con)
print(cla)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

linear_model = LinearRegression()
linear_model.fit(x_train, y_train)

# Make predictions on the test set
y_pred = linear_model.predict(x_test)
acc=accuracy_score(y_test,y_pred.round())
# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Display results
print(f'Mean Squared Error: {mse:.2f}')
print(f'R-squared: {r2:.2f}')
print("accuracy",acc)

# Plotting predictions against actual values
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Status")
plt.ylabel("Predicted Status")
plt.title("Linear Regression: Actual vs. Predicted Status")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create and train the Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(x_train, y_train)

# Make predictions on the test set
y_pred = rf_model.predict(x_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

# Display results
print(f'Accuracy: {accuracy:.2f}')
print(f'Confusion Matrix:\n{conf_matrix}')
print(f'Classification Report:\n{classification_rep}')
```
